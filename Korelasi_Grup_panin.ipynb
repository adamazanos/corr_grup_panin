#Korelasi Grup panin
Project ini bertujuan untuk mencari nilai korelasi dari harga saham Panin Grup yang sudah listing di bursa
Grup tersebut terdiri dari
PNBN , PNLF , PNIN , PANS , PNBS , AMAG , CFIN , VRNA , KIJA

```python
import yfinance as yf
import pandas as pd
import datetime
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=1000)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

#tarik data

```python
data1 = yf.download('PNBN.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data1["Date"] = data1.index
data1 = data1[["Close"
             , "Volume"]]
data1.reset_index(drop=True, inplace=True)
```

```python
data2 = yf.download('PNLF.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data2["Date"] = data2.index
data2 = data2[["Close"
             , "Volume"]]
data2.reset_index(drop=True, inplace=True)
```

```python
data3 = yf.download('PNIN.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data3["Date"] = data3.index
data3 = data3[["Close"
             , "Volume"]]
data3.reset_index(drop=True, inplace=True)
```

```python
data4 = yf.download('PANS.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data4["Date"] = data4.index
data4 = data4[["Close"
             , "Volume"]]
data4.reset_index(drop=True, inplace=True)
```

```python
data5 = yf.download('PNBS.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data5["Date"] = data5.index
data5 = data5[["Close"
             , "Volume"]]
data5.reset_index(drop=True, inplace=True)
```

```python
data6 = yf.download('AMAG.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data6["Date"] = data6.index
data6 = data6[["Close"
             , "Volume"]]
data6.reset_index(drop=True, inplace=True)
```

```python
data7 = yf.download('CFIN.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data7["Date"] = data7.index
data7 = data7[["Close"
             , "Volume"]]
data7.reset_index(drop=True, inplace=True)
```

```python
data8 = yf.download('VRNA.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data8["Date"] = data8.index
data8 = data8[["Close"
             , "Volume"]]
data8.reset_index(drop=True, inplace=True)
```

```python
data9 = yf.download('KIJA.JK',
                      start=start_date,
                      end=end_date,
                      progress=False)
data9["Date"] = data9.index
data9 = data9[["Close"
             , "Volume"]]
data9.reset_index(drop=True, inplace=True)
```

```python
data1.rename(columns = {'Close' : 'close_pnbn' ,'Volume':'vol_pnbn'}, inplace = True)
data2.rename(columns = {'Close' : 'close_pnlf' ,'Volume':'vol_pnlf'}, inplace = True)
data3.rename(columns = {'Close' : 'close_pnin' ,'Volume':'vol_pnin'}, inplace = True)
data4.rename(columns = {'Close' : 'close_pans' ,'Volume':'vol_pans'}, inplace = True)
data5.rename(columns = {'Close' : 'close_pnbs' ,'Volume':'vol_pnbs'}, inplace = True)
data6.rename(columns = {'Close' : 'close_amag' ,'Volume':'vol_amag'}, inplace = True)
data7.rename(columns = {'Close' : 'close_cfin' ,'Volume':'vol_cfin'}, inplace = True)
data8.rename(columns = {'Close' : 'close_vrna' ,'Volume':'vol_vrna'}, inplace = True)
data9.rename(columns = {'Close' : 'close_kija' ,'Volume':'vol_kija'}, inplace = True)
```
